```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
student = pd.read_excel("python-excel-1.xlsx")
student
```

```python
student.plot(kind="bar",x="Student Name",y="MTH 102",color="g")
```

```python
student.plot(kind="bar",x="Student Name",y=["MTH 102","PHY 192"],color=["g","r"])
```

```python
#plotting bar chart for bio 101
#figsize is used to increase the height and width
student.plot (kind="barh",x="Student Name",y="BIO 101",color="hotpink",
              figsize=(9,6))
plt.title("BAR CHART FOR BIOLOGY SCORE")
plt.show()
```

```python
#plotting LINE chart for CIT 142
#figsize is used to increase the height and width
student.plot (kind="line",x="Student Name",y="CIT 142",color="g",
              figsize=(12,6))
plt.title("LINE CHART FOR CIT SCORE")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(student["BIO 101"],labels=student["Student Name"])
plt.title("PIE CHAT FOR BIOLOGY SCORE")
plt.show()
```
